**Importing the necessary libraries**

```
import pandas as pd
import os
from datetime import datetime, timedelta
import math
import collections
import scipy.io
import os
```

**Changing given data to csv format from .mat**

```
format_matlab = scipy.io.loadmat(os.getcwd() + '/InsulinGlucoseData2.mat')
format_matlab = {x:y for x, y in format_matlab.items() if x[0] != '_'}
convertedNew = pd.DataFrame({x: pd.Series(y[0]) for x, y in format_matlab.items()})
convertedNew.to_csv(os.getcwd() + "/data.csv")
```

**Reading the converted CSV file**

```
data = pd.read_csv(os.getcwd() + '/data.csv')
```

```
dat_medicine = data['dateMuBolus'].dropna().tolist()
dat_CGM = data['dateNumber'].tolist()
value_medicine = data['actBolusDelivered'].dropna().tolist()
value_CGM = data['numCGM'].tolist()
```

**Converting String to Date Format**

```
def for_con(dates):
  for i in range(len(dates)):
    if not math.isnan(dates[i]):
      dates[i] = datetime.fromordinal(int(dates[i])) + timedelta(days=dates[i]%1) - timedelta(days = 366)
  return dates
```

```
date_CGM = for_con(dat_CGM)
date_medicine = for_con(dat_medicine)
```

```
print("CGM Dates", date_CGM[0:10])
print("Bolus Dates", date_medicine[0:10])
```

**Synchronizing the Dates for CGM and its values and Bolus and its values**


```
def nearest(items, pivot):
  return min(items, key=lambda x: abs(x - pivot))
```

```
assert len(date_CGM) == len(value_CGM)
assert len(date_medicine) == len(value_medicine)

DtoC = dict()
for i in range(len(date_CGM)):
  DtoC[date_CGM[i]] = value_CGM[i]

DtoB = dict()
for i in range(len(date_medicine)):
  DtoB[date_medicine[i]] = value_medicine[i]
```

```
DtoD = dict()
print(len(date_medicine))
for i in range(len(date_medicine)):
  print(i)
  dateNear = nearest(date_CGM, date_medicine[i])
  DtoD[date_medicine[i]] = dateNear
```

```
DtoD
```

**Saving the Synchronized Data**

```
data_frame = pd.DataFrame(columns=["DateCGM", "Value CGM", "DateBolus", "Value Bolus"])

kys_date = list(DtoD.keys())
date1_CGM = []
value1_CGM = []
date1_medicine = []
value1_medicine = []
for i in range(len(kys_date)):
  date1_CGM.append(kys_date[i])
  value1_CGM.append(DtoC[DtoD[kys_date[i]]])
  date1_medicine.append(DtoD[kys_date[i]])
  value1_medicine.append(DtoB[kys_date[i]])

data_frame['DateCGM'] = date1_CGM
data_frame['Value CGM'] = value1_CGM
data_frame['DateBolus'] = date1_medicine
data_frame['Value Bolus'] = value1_medicine

data_frame = data_frame.dropna()
data_frame.to_csv('Final_Data.csv', index=False)
```
